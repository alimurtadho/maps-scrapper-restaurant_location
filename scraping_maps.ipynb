```python
#cara ke dua

import pandas as pd, numpy as np
import requests
import json
import time

final_data=[]

# Parameters
datakecamatanbdg = pd.read_csv('YOUR_FILE_NAME.CSV', sep=';')
kecamatanbdg = datakecamatanbdg.stack().str.replace(' ', '+').unstack().set_index("Kecamatan")
api_key = 'AIzaSyAiFpFd85eMtfbvmVNEYuNds5TEF9FjIPI' #insert your Places API

for fix_kecamatan in kecamatanbdg.index:
    url = 'https://maps.googleapis.com/maps/api/place/textsearch/json?query=tempat+ngopi+di+kecamatan+' + str(
        fix_kecamatan) + '&key=' + str(api_key)
    print(url)
    respon = requests.get(url)
    jj = json.loads(respon.text)
    results = jj['results']
    for result in results:
        name = result['name']
        place_id = result['place_id']
        lat = result['geometry']['location']['lat']
        lng = result['geometry']['location']['lng']
        rating = result['rating']
        types = result['types']
        total_ratings = result['user_ratings_total']
        address = result['formatted_address']
        data = [name, place_id, lat, lng, rating, types, total_ratings, address]
        final_data.append(data)
    time.sleep(5)

labels = ['Place Name', 'Place ID', 'Latitude', 'Longitude', 'Rating', 'Types', 'Total Ratings', 'Address']
fix_data = pd.DataFrame.from_records(final_data, columns=labels).drop_duplicates('Place Name').set_index('Place Name')
fix_data.to_csv('Tempat_Ngopi_Bandung.csv')
```
